```python
import numpy as np
import pandas as pd
import random
```

```python
df = pd.read_csv("./data/diabetes/diabetes_012_health_indicators_BRFSS2015.csv")
df.head()
```

```python
target_rem = df.drop(["Diabetes_012"], axis=1)

row, col = np.where(np.random.random(target_rem.shape) < .001)
for r,c in zip(row, col):
    target_rem.iat[r, c] = np.nan
```

```python
final_df = pd.concat([df["Diabetes_012"], target_rem], axis=1)
final_df.isna().sum()
```

```python
final_df.to_csv("./data/diabetes/diabetes_data_with_missing.csv", index=False)
```
